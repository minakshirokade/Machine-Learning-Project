```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
```

```python
#Step1 : to create dataset
data = {
    'Cheap': ['Yes', 'Yes', 'No', 'Yes', 'No'],
    'Promo': ['Yes', 'No', 'Yes', 'Yes', 'No'],
    'Buy': ['Yes', 'Yes', 'No', 'Yes', 'No'],
}
    
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Step 2:Encode categorical variables
le_cheap= LabelEncoder()
le_promo = LabelEncoder()
le_buy = LabelEncoder()
```

```python
df['Cheap'] = le_cheap.fit_transform(df['Cheap'])   #Yes = 1, No = 0
df['Promo'] = le_promo.fit_transform(df['Promo']) 
df['Buy'] = le_buy.fit_transform(df['Buy']) 
```

```python
# Step3: Features(X) and target(Y)
x  = df[['Cheap', 'Promo']]
y = df['Buy']
```

```python
#Step 4:Train Naive Bayes 
model = CategoricalNB()
model.fit(x,y)
```

```python
#Step 5: To predict the same data
predictions = model.predict(x)
```

```python
#Step 6: Show results with decoded labels
df['Predicted_Buy'] = le_buy.inverse_transform(predictions)
df['Actual_Buy'] = le_buy.inverse_transform(y)

```

```python
data1= {
    'Weather' : ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain'],
    'Wind' : ['Weak', 'Strong', 'Weak', 'Weak', 'Strong'],
    'Play' : ['Yes', 'No', 'Yes', 'Yes', 'No']
}
```

```python
df= pd.DataFrame(data1)
```

```python
le_Weather = LabelEncoder()
le_Wind = LabelEncoder()
le_Play = LabelEncoder()
```

```python
df['Weather'] = le_Weather.fit_transform(df['Weather']) #Sunny=1, Overcast=2, Rain = 3
df['Wind'] = le_Wind.fit_transform(df['Wind']) #Weak = 0, Strong = 1
df['Play'] = le_Play.fit_transform(df['Play']) #Yes = 0, No = 0
```

```python
x = df[['Weather', 'Wind']]
y = df['Play']
```

```python
model =  CategoricalNB()
model.fit(x,y)
```

```python
prediction = model.predict(x)
```

```python
df['Predicted_Play'] = le_Play.inverse_transform(prediction)
df['Actual_Play'] = le_Play.inverse_transform(prediction)
```
